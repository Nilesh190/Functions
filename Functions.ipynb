5. What are iterators in Python and how do they differ from iterables?



*   Iterable:

1.   An object that can be looped over (like list, tuple, string, etc.).
2.   It implements the __iter__() method, It does not keep track of where it is during iteration.

*    Iterator:

1.  An object that represents a stream of data.
2.  It remembers the current position
3.  It implements two methods:

__iter__() → returns the iterator object itself

__next__() → returns the next item or raises StopIteration






```python
nums = [1, 2, 3]  # This is an iterable

for num in nums:
    print(num)

```

```python
nums = [1, 2, 3]
it = iter(nums)  # Now 'it' is an iterator

print(next(it))  # Output: 1
print(next(it))  # Output: 2
print(next(it))  # Output: 3
# print(next(it))  # Raises StopIteration
```

6. Explain the concept of generators in Python and how they are defined.

A generator is a special type of iterator in Python that yields values one at a time, only when needed, using the yield keyword.


*   Defined like functions but use yield instead of return.
*   Do not store all values in memory — memory efficient.

*   Automatically create an iterator when called.
*   Used for lazy evaluation





```python
def count_up_to_three():
    yield 1
    yield 2
    yield 3

gen = count_up_to_three()

for num in gen:
    print(num)
```

7. What are the advantages of using generators over regular functions?



*    Memory Efficiency: Generators use lazy evaluation, meaning they yield one value at a time and do not store the entire output in memory. This is ideal when dealing with large data or streams.
*   Faster Execution : Generators start producing
values immediately when called, unlike regular functions that compute the entire output before returning. This reduces initial load time and improves performance in many cases.


*   Can Represent Infinite Sequences: Because generators yield values one at a time and maintain state, they can be used to model infinite or very large sequences without running out of memory.
*   Clean and Simple Syntax: Generators simplify code when handling sequences. Instead of managing state manually, yield automatically pauses and resumes the function, making the logic easier to read and write.




```python
def count_forever():
    num = 1
    while True:
        yield num
        num += 1

gen = count_forever()
for i in range(5):
    print(next(gen))
```

9.  Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new map object (which is an iterator).


Purpose:
*   To transform each element in a collection using a function.
*   Avoids explicit loops, making code shorter and cleaner.






```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
result = map(square, numbers)

print(list(result))  # Output: [1, 4, 9, 16]
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

1.   map() Function
*   Purpose: Applies a function to each element in an iterable.
*   Returns: A new iterable (map object) with transformed values.
2.   filter() Function
*   Purpose: Filters elements based on a condition (True/False).
*   Returns: A new iterable (filter object) with only matching elements.
3.  reduce() Function
*   Purpose: Repeatedly applies a function to the elements, reducing them to a single value.
*   Returns: One final result.







```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Step 1: map() → square each number
squared = list(map(lambda x: x * x, numbers))
print("Squared:", squared)  # [1, 4, 9, 16, 25]

# Step 2: filter() → keep even squares
even_squares = list(filter(lambda x: x % 2 == 0, squared))
print("Even Squares:", even_squares)  # [4, 16]

# Step 3: reduce() → sum the even squares
sum_even_squares = reduce(lambda x, y: x + y, even_squares)
print("Sum of Even Squares:", sum_even_squares)  # 20
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):

  even_sum = 0
  for number in numbers:
    if number % 2 == 0:
      even_sum += number
  return even_sum
```

4.  Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def is_prime(n):
    if n <= 1:
        return False  # 1 and below are not prime
    for i in range(2, int(n**0.5) + 1):  # check up to square root
        if n % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a Prime number")
    else:
        print(num, "is NOT a Prime number")
```

5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        if num_terms < 0:
            raise ValueError("Number of terms must be a non-negative integer.")
        self._num_terms = num_terms
        self._count = 0
        self._a = 0
        self._b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self._count < self._num_terms:
            if self._count == 0:
                self._count += 1
                return self._a
            elif self._count == 1:
                self._count += 1
                return self._b
            else:
                next_fib = self._a + self._b
                self._a = self._b
                self._b = next_fib
                self._count += 1
                return next_fib
        else:
            raise StopIteration

# Example usage:
fib_iter = FibonacciIterator(10)
print("Fibonacci sequence up to 10 terms:")
for num in fib_iter:
    print(num, end=" ")
print()

```
